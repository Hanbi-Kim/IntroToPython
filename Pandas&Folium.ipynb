## 실습: Folium을 활용하여 인구데이터 시각화하기

```python
import folium
import json
import re

# 서울시 구별 지도 정보가 포함된 geo_json 파일
geo_json='https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'

# folium이란 패키지를 가져와 지도를 형성
m=folium.Map(
    location=[37.566345,126.977893], #기준이되는 경위도 정보 서울시= [37.566345,126.977893]
    tiles='OpenStreetMap' # 지도 종류 'OpenStreetMap' / 'Stamen Terrain' 
)

# 특정 부분에 마커 표시하기.
folium.Marker([37.566345,126.977893],
              popup="서울시 기준",
              tooltip="서울시 기준이 되는 곳").add_to(m)
m
m
```

```python

# 서울시 구별 지도 정보가 포함된 geo_json 파일
geo_json='https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'

# folium이란 패키지를 가져와 지도를 형성
m=folium.Map(
    location=[37.566345,126.977893], #기준이되는 경위도 정보 서울시= [37.566345,126.977893]
    tiles='OpenStreetMap', # 지도 종류 'OpenStreetMap' / 'Stamen Terrain' 
    zoom_start = 11 # 시작시 줌 위치 커질수록 가깝게
)

df2=df[['행정구역별','계']]
df2.columns=['name','values'] # geo_json 파일에 행정구역별 이름이 name으로 저장되어있음

folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=df2,
    columns=['name','values'],
    key_on='feature.properties.name',
    fill_color='BuGn', #타일색 종류 ['BuGn', 'BuPu', 'GnBu', 'OrRd', 'PuBu', 'PuBuGn', 'PuRd', 'RdPu', 'YlGn', 'YlGnBu', 'YlOrBr', 'YlOrRd' ]
    fill_opacity=0.9, #타일투명도
    line_opacity=0.1 #선투명도
).add_to(m)

m
```
